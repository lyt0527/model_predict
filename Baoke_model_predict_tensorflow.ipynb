```python
import pandas as pd
import numpy
import time
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```python
start = time.time()
data = pd.read_csv("/srv/BaseLine/GuarantorForecast/data/user_info_complain_v1.csv")
print(time.time() - start)
data
```

```python
data.columns
```

```python
data = data[data.type.fillna('0').astype('int') < 2]
```

```python
# 根据身份证筛选取长度为18，第7到11位和倒数第二位为数字 297980
data = data[(data.id_number.str.len() == 18) & (data.id_number.str[6:10].str.isdigit()) & data.id_number.str[-2: -1].str.isdigit()] 

data['gender'] = data.id_number.str[-2: -1].map(lambda x: 1 if int(x) % 2 == 1 else 0)
data['age'] = 2020 - data.id_number.str[6:10].astype('int')
data['years'] = data['years'].fillna(data.years.mean())
data['car_system'] = data['car_system'].fillna('0')
data['amount_money'] = data['amount_money'].fillna(data.amount_money.mean())
data['amount_repair_times'] = data['amount_repair_times'].fillna('0')
data['is_repair'] = data['is_repair'].map(lambda x : 1 if x == '大修' else 0)
data['is_driver'] = data['is_driver'].map(lambda x : 1 if x == '试驾' else 0)
data['is_complain'] =data['is_complain'].fillna('0')
data['type'] = data['type'].fillna('0')
filter_data = data.loc[:, ['gender', 'age', 'years', 'amount_money', 'amount_repair_times', 'is_repair', 'is_driver', 'is_complain', 'type']]
filter_data
```

```python
filter_data = filter_data[(filter_data.age >= 18) & (filter_data.age <= 100)]
df = filter_data[filter_data.type.astype('int') < 2]
```

```python
df
```

```python
ages = pd.cut(df.age, 10)
age = pd.get_dummies(ages)
age.fillna(0, inplace=True)

gender = pd.get_dummies(df.gender)
gender.fillna(0, inplace=True)

years = pd.cut(df.years, 10)
years = pd.get_dummies(years)
years.fillna(0, inplace=True)


amount_money = pd.cut(df.amount_money, 10)
amount_money = pd.get_dummies(amount_money)
amount_money.fillna(0, inplace=True)

amount_repair_times = pd.cut(df.amount_repair_times.astype('int'), 50)
amount_repair_times = pd.get_dummies(amount_repair_times)
amount_repair_times.fillna(0, inplace=True)

is_repair = pd.get_dummies(df.is_repair)
is_repair.fillna(0, inplace=True)

is_driver = pd.get_dummies(df.is_driver)
is_driver.fillna(0, inplace=True)

print(age.shape, gender.shape, years.shape, amount_money.shape, amount_repair_times.shape, is_repair.shape, is_driver.shape)
```

```python
# buy_type = df.type
buy_type = df.pop('type')
```

```python
df = pd.concat([gender, age, years, amount_money, amount_repair_times, is_repair, is_driver], axis=1, ignore_index=True)
df
```

```python
X = np.array(df)
X.shape
```

```python
y = np.array(buy_type)
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape
```

```python
model = tf.keras.Sequential([
    layers.Dense(32, activation='sigmoid',  input_shape=(86, ), kernel_regularizer=tf.keras.regularizers.l2(0.001)),
#     layers.Dropout(0.5),
    layers.Dense(16, activation='relu'),
    layers.Dense(1, activation='relu')
])
model.summary()
```

```python
model.compile(loss='mse', optimizer='Adam', metrics=['accuracy'])
model.fit(X_train, y_train, batch_size=128, epochs=20, validation_split=0.2, shuffle=True)
```

```python
loss, accuracy = model.evaluate(X_train, y_train)
print(loss)
print(accuracy)
```

```python
Y_pred = model.predict(X_test)
Y_pred
```

```python
r2_score(y_test, Y_pred)
```
